```python
import sys
import torch
from torch.utils.data import DataLoader

# Add root directory to path to allow importing from src/ without installing as a package
sys.path.append('../') 

from src.data.dataset import LFWTrainingDataset, PKBatchSampler
from src.data.hard_pair_mining import HardPairMiner
```

```python
# 1. Initialize Dataset and custom Sampler

# For testing purposes, we can use a smaller batch size. In practice, we would want to use a larger batch size 
# (e.g., P=16, K=4 -> Batch Size = 64) for better training stability and performance.
dataset = LFWTrainingDataset(processed_data_dir='../data/processed', img_size='S', min_images=3)

# The PKBatchSampler will create batches of size P*K, where each batch contains P unique identities and K images per identity.
sampler = PKBatchSampler(dataset=dataset, p=4, k=4)
dataloader = DataLoader(dataset, batch_sampler=sampler)
```

```python
# 2. Fetch one batch
images, labels = next(iter(dataloader))
print(f"Batch images shape: {images.shape}") # Expected: [16, 3, 300, 300]
print(f"Batch labels:\n{labels}\n")
```

```python
# 3. Simulate backbone embeddings and test the Miner
# Create 16 dummy embeddings of size 1280
dummy_embeddings = torch.randn(16, 1280) 

miner = HardPairMiner()
hard_pos_idx, hard_neg_idx = miner(dummy_embeddings, labels)

print("Miner Output:")
print(f"Hard Positive Indices:\n{hard_pos_idx}")
print(f"Hard Negative Indices:\n{hard_neg_idx}")
```
